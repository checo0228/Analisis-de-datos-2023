Graficar los valores globales y los de Mexico para: 'Total', 'Coal', 'Oil', 'Gas', 'Cement', 'Flaring', 'Per Capita'.

```python
df_mexico = df[df['Country'] == 'Mexico']
df_mundo = df[df['Country'] == 'Global']

# Crear una lista de las categorías a graficar
categorias = ['Total', 'Coal', 'Oil', 'Gas', 'Cement', 'Flaring', 'Per Capita']

# Crear una figura y varios ejes (subplots) en una sola llamada
fig, axs = plt.subplots(nrows=3, ncols=3, figsize=(12, 8))

# Aplanar la matriz de ejes en una lista para facilitar la iteración
axs = axs.flatten()

# Crear un gráfico para cada categoría
for i, categoria in enumerate(categorias):
    # Graficar los valores globales
    axs[i].plot(df_mundo['Year'], df_mundo[categoria], label='Global')
    # Graficar los valores de México
    axs[i].plot(df_mexico['Year'], df_mexico[categoria], label='México')
    # Configurar el gráfico
    axs[i].set_title(categoria)
    axs[i].set_xlabel('Año')
    axs[i].set_ylabel('MtCO2')
    axs[i].legend()

# Ajustar los espacios entre los subplots
plt.tight_layout()

# Mostrar los gráficos
plt.show()
```

Extraer datos de los países pertenecientes al G20 y graficar las columnas del punto anterior.

```python

# Cargar los datos del archivo csv
df = pd.read_csv('dataset.csv')

# Seleccionar los países del G20
g20 = ['Argentina', 'Australia', 'Brazil', 'Canada', 'China', 'France', 'Germany', 'India',
       'Indonesia', 'Italy', 'Japan', 'Mexico', 'Russia', 'Saudi Arabia', 'South Africa',
       'South Korea', 'Turkey', 'United Kingdom', 'United States']

# Filtrar los datos para incluir solo los países del G20
df_g20 = df[df['Country'].isin(g20)]

# Seleccionar las categorías a graficar
categorias = ['Total', 'Coal', 'Oil', 'Gas', 'Cement', 'Flaring', 'Per Capita']

# Crear una figura con 7 subplots
fig, axs = plt.subplots(nrows=1, ncols=7, figsize=(20, 5))

# Recorrer las categorías y graficar los valores globales y del G20
for i, categoria in enumerate(categorias):
    # Obtener los valores globales
    datos_mundo = df.groupby('Year')[categoria].sum()
    # Obtener los valores del G20 (excluyendo el valor 'Global')
    datos_g20 = df_g20[df_g20['Country'] != 'Global'].groupby('Year')[categoria].sum()
    # Graficar los valores globales
    axs[i].plot(df['Year'].unique(), datos_mundo, label='Mundo')
    # Graficar los valores del G20
    axs[i].plot(datos_g20.index, datos_g20, label='G20')
    # Configurar el título y las etiquetas de los ejes
    axs[i].set_title(categoria)
    axs[i].set_xlabel('Año')
    axs[i].set_ylabel('Millones de toneladas de CO2')
    # Agregar una leyenda a la figura
    axs[i].legend()

# Mostrar la figura
plt.show()
```

Cuales son los 3 países que más contaminan en cada uno de los casos

```python
# Calcular los 3 países que más contaminan en cada categoría
for categoria in categorias:
    datos_categoría = df_g20.groupby('Country')[categoria].sum().sort_values(ascending=False)
    top_3 = datos_categoría.head(3)
    print(f'Los 3 países que más contaminan en {categoria} son:')
    print(top_3)
```

Considerando solo la emisión de los paises del G20, con que porcentaje contribuye México al Total y Per Capita?

```python
# Extraer los países del G20
paises_g20 = ['Argentina', 'Australia', 'Brazil', 'Canada', 'China', 'France', 'Germany',
              'India', 'Indonesia', 'Italy', 'Japan', 'Mexico', 'Russia', 'Saudi Arabia',
              'South Africa', 'South Korea', 'Turkey', 'United Kingdom', 'United States']

# Filtrar los datos del G20
df_g20 = df[df['Country'].isin(paises_g20)]

# Graficar las columnas seleccionadas para cada país del G20
categorias = ['Total', 'Coal', 'Oil', 'Gas', 'Cement', 'Flaring', 'Per Capita']
for categoria in categorias:
    # Crear una figura y ejes para la gráfica
    fig, ax = plt.subplots(figsize=(8, 6))
    # Graficar los valores de cada país del G20
    for pais in paises_g20:
        datos_pais = df_g20[df_g20['Country'] == pais].groupby('Year')[categoria].sum()
        ax.plot(datos_pais.index, datos_pais, label=pais)
    # Configurar la gráfica
    ax.set_xlabel('Año')
    ax.set_ylabel(categoria + ' (MtCO2)')
    ax.set_title('Emisiones de ' + categoria + ' en los países del G20')
    ax.legend(loc='upper left')
    plt.show()
```
